```python
spark.sql("DROP TABLE IF EXISTS tmp_hmg_tipo_servicio")

spark.sql("""
    CREATE OR REPLACE TABLE tmp_hmg_tipo_servicio AS
    SELECT * FROM tfm_silver_hmg.homologacion WHERE dominio = 'tipo_servicio' LIMIT 0
""")
```

```python
spark.sql("""
    INSERT INTO tmp_hmg_tipo_servicio (dominio, columna_origen, valor_original, catalog_id, fecha_desde, fecha_hasta) VALUES
    ('tipo_servicio', 'TIPO_SERVICIO', 'ALQ', 3, '1900-01-01 00:00:00' , NULL ),
    ('tipo_servicio', 'TIPO_SERVICIO', 'PAR', 2, '1900-01-01 00:00:00' , NULL );
""")
```

```python
%sql
INSERT INTO tfm_silver_hmg.homologacion (dominio, columna_origen, valor_original, catalog_id, fecha_desde, fecha_hasta)
SELECT t.dominio, t.columna_origen,t.valor_original, t.catalog_id ,t.fecha_desde, t.fecha_hasta
FROM tmp_hmg_tipo_servicio t
LEFT JOIN tfm_silver_hmg.homologacion c
    ON t.dominio = c.dominio
    AND t.columna_origen = c.columna_origen
    AND t.valor_original = c.valor_original
    AND t.catalog_id = c.catalog_id
WHERE c.regla_id IS NULL
```
