```python
spark.sql("DROP TABLE IF EXISTS tmp_cat_clase")

spark.sql("""
    CREATE OR REPLACE TABLE tmp_cat_clase AS
    SELECT * FROM tfm_silver_ctg.cat_clase LIMIT 0
""")
```

```python
%sql
INSERT INTO tfm_silver_ctg.cat_clase (clase, descripcion, fecha_desde, fecha_hasta)
SELECT t.clase, t.descripcion, t.fecha_desde, t.fecha_hasta
FROM tmp_cat_clase t
LEFT JOIN tfm_silver_ctg.cat_clase c
  ON t.clase = c.clase
WHERE c.clase IS NULL
```
