```python
%sql
-- Dimensiones GOLD (si no existe)
CREATE DATABASE IF NOT EXISTS tfm_gold
COMMENT 'Dimensiones de la capa GOLD';
```

```python
%sql

-- Dimensión de cantones en GOLD 
CREATE TABLE IF NOT EXISTS tfm_gold.dim_canton (
  ID_CANTON              BIGINT,
  DESC_CANTON            STRING,
  ID_PROVINCIA           INT,
  DESC_PROVINCIA         STRING,
  ID_REGION              INT,
  DESC_REGION            STRING,
  PAIS                   STRING,
  LATITUD                DOUBLE,
  LONGITUD               DOUBLE,
  FECHA_DESDE            TIMESTAMP,
  FECHA_HASTA            TIMESTAMP,
  CANTON_PROVINCIA_PAIS  STRING  
)
USING DELTA
COMMENT 'Dimensión de cantones (GOLD)'
TBLPROPERTIES (
  delta.autoOptimize.optimizeWrite = true,
  delta.autoOptimize.autoCompact   = true
);
```
