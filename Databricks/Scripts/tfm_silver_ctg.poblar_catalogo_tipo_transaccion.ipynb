```python
spark.sql("DROP TABLE IF EXISTS tmp_cat_tipo_transaccion")

spark.sql("""
    CREATE OR REPLACE TABLE tmp_cat_tipo_transaccion AS
    SELECT * FROM tfm_silver_ctg.cat_tipo_transaccion LIMIT 0
""")
```

```python
%sql
INSERT INTO tfm_silver_ctg.cat_tipo_transaccion (tipo_transaccion, descripcion, fecha_desde, fecha_hasta)
SELECT t.tipo_transaccion, t.descripcion, t.fecha_desde, t.fecha_hasta
FROM tmp_cat_tipo_transaccion t
LEFT JOIN tfm_silver_ctg.cat_tipo_transaccion c
  ON t.tipo_transaccion = c.tipo_transaccion
WHERE c.tipo_transaccion IS NULL
```
