```python
%sql
CREATE DATABASE IF NOT EXISTS tfm_bronze
COMMENT 'Capa Bronce:';
```

```python
%sql
--DROP TABLE IF EXISTS tfm_bronze.state_files;
CREATE TABLE IF NOT EXISTS tfm_bronze.state_files (
  year INT COMMENT 'Año del dataset fuente. Representa la partición lógica correspondiente al archivo procesado en S3.',
  last_metadata_update TIMESTAMP COMMENT 'Fecha y hora de la última actualización registrada en los metadatos del archivo (extraída desde el portal de origen o el manifiesto).',
  last_status STRING COMMENT 'Estado actual del archivo en S3',
  last_checked_ts TIMESTAMP COMMENT 'Timestamp que indica cuándo fue verificado por última vez el estado del archivo en el proceso de monitoreo o ingesta.',
  last_file_path STRING COMMENT 'Ruta completa del archivo dentro del bucket S3 en la capa Raw. Permite rastrear el origen exacto de los datos.',
  last_sha256 STRING COMMENT 'Hash SHA-256 calculado sobre el contenido del archivo. Se utiliza para identificar cambios o duplicados de manera confiable.'
)
USING DELTA
COMMENT 'Tabla de control en la capa Bronze que almacena el estado, ruta y hash de los archivos CSV presentes en el bucket S3 Raw.';
```
