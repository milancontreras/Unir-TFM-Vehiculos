```python
%sql
-- =========================================================
-- Carga incremental de la dimensión de Cantones GOLD
-- Solo inserta los cantones nuevos que no existan aún
-- =========================================================
INSERT INTO tfm_gold.dim_canton
SELECT
  s.ID_CANTON,
  s.DESC_CANTON,
  s.ID_PROVINCIA,
  s.DESC_PROVINCIA,
  s.ID_REGION,
  s.DESC_REGION,
  s.PAIS,
  s.LATITUD,
  s.LONGITUD,
  s.FECHA_DESDE,
  s.FECHA_HASTA,
  CONCAT_WS(', ',
    NULLIF(TRIM(s.DESC_CANTON), ''),
    NULLIF(TRIM(s.DESC_PROVINCIA), ''),
    NULLIF(TRIM(s.PAIS), '')
  ) AS CANTON_PROVINCIA_PAIS
  FROM tfm_silver_dim.dim_canton s
--  Solo los que NO están en GOLD
LEFT ANTI JOIN tfm_gold.dim_canton g
  ON g.ID_CANTON = s.ID_CANTON;
```
