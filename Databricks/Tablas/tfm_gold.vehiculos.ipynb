```python
%sql
CREATE DATABASE IF NOT EXISTS tfm_gold
COMMENT 'Capa Oro - Data depurada y enriquecida';
```

```python
%sql
CREATE TABLE IF NOT EXISTS tfm_gold.vehiculos (
    ANIO_ORIGEN        INT,
    CODIGO_VEHICULO    STRING,
    TIPO_TRANSACCION   STRING,
    MARCA              STRING,
    MODELO             STRING,
    PAIS               STRING,
    ANIO_MODELO        INT,
    CLASE              STRING,
    SUBCLASE           STRING,
    TIPO_VEHICULO      STRING,
    AVALUO             DOUBLE,
    FECHA_PROCESO      DATE,
    TIPO_SERVICIO      STRING,
    CILINDRAJE         INT,
    TIPO_COMBUSTIBLE   STRING,
    FECHA_COMPRA       DATE,
    ID_CANTON          BIGINT,
    COLOR1             STRING,
    TIPO_COMPRADOR     STRING,
    INGEST_TS          TIMESTAMP
)
USING DELTA
PARTITIONED BY (ANIO_ORIGEN)
TBLPROPERTIES (
  delta.autoOptimize.optimizeWrite = true,
  delta.autoOptimize.autoCompact   = true
);
```
