```python
spark.sql("DROP TABLE IF EXISTS tmp_cat_tipo_comprador")

spark.sql("""
    CREATE OR REPLACE TABLE tmp_cat_tipo_comprador AS
    SELECT * FROM tfm_silver_ctg.cat_tipo_comprador LIMIT 0
""")
```

```python
spark.sql("""
    INSERT INTO tmp_cat_tipo_comprador (tipo_comprador, descripcion, fecha_desde, fecha_hasta) VALUES
    ('N/E', 'No especificado o Faltante', '1900-01-01 00:00:00', NULL),
    ('NATURAL'   , NULL, '1900-01-01 00:00:00', NULL),
    ('JURIDICA'   , NULL, '1900-01-01 00:00:00', NULL);
""")
```

```python
%sql
INSERT INTO tfm_silver_ctg.cat_tipo_comprador (tipo_comprador, descripcion, fecha_desde, fecha_hasta)
SELECT t.tipo_comprador, t.descripcion, t.fecha_desde, t.fecha_hasta
FROM tmp_cat_tipo_comprador t
LEFT JOIN tfm_silver_ctg.cat_tipo_comprador c
  ON t.tipo_comprador = c.tipo_comprador
WHERE c.tipo_comprador IS NULL
```
