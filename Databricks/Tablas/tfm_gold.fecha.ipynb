```python
%sql
CREATE DATABASE IF NOT EXISTS tfm_gold
COMMENT 'Dimensiones de la capa GOLD';
```

```python

%sql
CREATE TABLE IF NOT EXISTS tfm_gold.dim_fecha (
  Fecha              DATE,        -- clave de la dimensión (Date)
  Anio               INT,         -- YEAR([Date])
  MesNumero          INT,         -- MONTH([Date])
  MesNombre          STRING,      -- "MMMM" (ES)
  MesAbrev           STRING,      -- "MMM"  (ES)
  AnioMes            INT,         -- YYYYMM (para ordenar)
  Dia                INT,         -- DAY([Date])
  DiaSemanaNum       INT,         -- 1=Lunes ... 7=Domingo
  DiaSemanaNombre    STRING,      -- "dddd" (ES)
  Trimestre          STRING,      -- Q1..Q4
  Semestre           STRING       -- S1 / S2
)
USING DELTA
COMMENT 'Dimensión de Fechas generada desde tfm_gold.vehiculos'
TBLPROPERTIES (
  delta.autoOptimize.optimizeWrite = true,
  delta.autoOptimize.autoCompact   = true
);
```
