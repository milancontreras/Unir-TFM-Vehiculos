```python
%sql
CREATE DATABASE IF NOT EXISTS tfm_silver_dim
COMMENT 'Dimensiones de la capa PLATA';
```

```python
%sql

-- Dimensión de cantones
CREATE TABLE IF NOT EXISTS tfm_silver_dim.dim_canton (
  ID_CANTON       BIGINT       COMMENT 'Código único de cantón',
  DESC_CANTON     STRING       COMMENT 'Nombre del cantón',
  ID_PROVINCIA    INT          COMMENT 'Código de provincia',
  DESC_PROVINCIA  STRING       COMMENT 'Nombre de provincia',
  ID_REGION       INT          COMMENT 'Código de región',
  DESC_REGION     STRING       COMMENT 'Nombre de región',
  PAIS            STRING       COMMENT 'País ',
  LATITUD         DOUBLE       COMMENT 'Latitud (grados decimales)',
  LONGITUD        DOUBLE       COMMENT 'Longitud (grados decimales)',
  FECHA_DESDE     TIMESTAMP    COMMENT 'Fecha inicio vigencia',
  FECHA_HASTA     TIMESTAMP    COMMENT 'Fecha fin vigencia'
)
USING DELTA
COMMENT 'Dimension de cantones de clases vehiculos';
```
