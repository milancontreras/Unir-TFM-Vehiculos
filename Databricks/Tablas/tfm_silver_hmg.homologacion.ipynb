```python
%sql
CREATE DATABASE IF NOT EXISTS tfm_silver_hmg
COMMENT 'Homologaciones Capa Plata';
```

```python
%sql
--DROP TABLE IF EXISTS tfm_silver_hmg.homologacion;
CREATE TABLE IF NOT EXISTS tfm_silver_hmg.homologacion (
    regla_id            BIGINT GENERATED ALWAYS AS IDENTITY,  
    dominio             STRING,        
    columna_origen      STRING,        
    valor_original      STRING,        
    catalog_id          BIGINT,        -- referencia al catálogo (puede ser NULL al inicio)
    fecha_desde         TIMESTAMP,
    fecha_hasta         TIMESTAMP
)
USING DELTA
COMMENT 'Reglas de homologación de valores provenientes de Bronze hacia valores normalizados/catalogados';
```
